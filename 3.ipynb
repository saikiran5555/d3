For a square matrix to be diagonalizable using the Eigen-Decomposition approach, it must satisfy the following conditions:

The matrix must be square: A matrix must have an equal number of rows and columns to be diagonalizable.

The matrix must have n linearly independent eigenvectors: For an n x n matrix A to be diagonalizable, it must have n linearly independent eigenvectors. This condition ensures that the eigenvectors span the entire vector space, allowing for a complete decomposition of the matrix.

The matrix must have a full set of eigenvalues: A square matrix must have a full set of n eigenvalues (not necessarily distinct) to be diagonalizable. Each eigenvalue corresponds to a linearly independent eigenvector, and together they form the basis for the diagonalization process.
